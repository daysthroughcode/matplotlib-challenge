## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_md = pd.read_csv(mouse_metadata_path)
study_res = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study = pd.merge(mouse_md, study_res, how ='inner', on ='Mouse ID')


# Display the data table for preview
mouse_study
```

```python
mouse_study.shape
```

```python
# Checking the number of mice.

#Sort by Mouse ID then Timepoint

mouse_study.sort_values(['Mouse ID','Timepoint'], ascending = False) 

#Count and Call Unique values

mouse_num = mouse_study['Mouse ID'].nunique()
mouse_num
```

```python
#Modify data to use Mouse ID as index

mouse_index = mouse_study.set_index("Mouse ID", drop = False)
mouse_index
```

```python
#Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicatemice = mouse_study[mouse_study.duplicated()]
duplicatemice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

mouse_index.loc["g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_indcln = mouse_index.drop('g989')
mouse_indcln.shape
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_numcln = mouse_indcln['Mouse ID'].nunique()
mouse_numcln
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mse_mean = round(mouse_indcln["Tumor Volume (mm3)"].mean(),2)
mse_median = round(mouse_indcln["Tumor Volume (mm3)"].median(),2)
mse_var = round(mouse_indcln["Tumor Volume (mm3)"].var(),2)
mse_sd = round(mouse_indcln["Tumor Volume (mm3)"].std(),2)
mse_sem = round(mouse_indcln["Tumor Volume (mm3)"].sem(),2)

sumstat1 = dict()
sumstat1['Mean'] = str(mse_mean)
sumstat1['Median'] = str(mse_median)
sumstat1['Variance'] = str(mse_var)
sumstat1['Standard Deviation'] = str(mse_sd)
sumstat1['Standard Error of Mean'] = str(mse_sem)

sumstat1df = pd.DataFrame(list(sumstat1.items()), index=[' ', ' ', ' ', ' ', ' '], columns =('Summary Statistic','Value'))
# This method is the most straighforward, creating multiple series and putting them all together at the end.

sumstat1df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

sumstat2 = st.describe(mouse_indcln["Tumor Volume (mm3)"])
sumstat2

# This method produces everything in a single groupby function
```
